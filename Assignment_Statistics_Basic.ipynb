```python
#7 Calculate skewness and kurtosis of a dataset using Python libraries
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
import seaborn as sns

# Positively Skewed Dataset (Right-skewed) - Exponential Distribution
pos_data = np.random.exponential(scale=2.0, size=1000)

# Negatively Skewed Dataset (Left-skewed)
neg_data = -np.random.exponential(scale=2.0, size=1000)

# Calculate skewness and kurtosis for the positively skewed data
pos_skewness = skew(pos_data)
pos_kurtosis = kurtosis(pos_data, fisher=False)

# Calculate skewness and kurtosis for the negatively skewed data
neg_skewness = skew(neg_data)
neg_kurtosis = kurtosis(neg_data, fisher=False)

# Print results
print(f"Positively Skewed Data: Skewness = {pos_skewness:.2f}, Kurtosis = {pos_kurtosis:.2f}")
print(f"Negatively Skewed Data: Skewness = {neg_skewness:.2f}, Kurtosis = {neg_kurtosis:.2f}")

# Plotting histograms for visual representation
plt.figure(figsize=(12, 6))

# Positively skewed data plot
plt.subplot(1, 2, 1)
sns.histplot(pos_data, bins=30, kde=True, color='skyblue')
plt.title("Positively Skewed Data")

# Negatively skewed data plot
plt.subplot(1, 2, 2)
sns.histplot(neg_data, bins=30, kde=True, color='salmon')
plt.title("Negatively Skewed Data")

plt.tight_layout()
plt.show()
```

```python
#9  Write a Python script to calculate covariance between two datasets
import numpy as np

# Sample datasets
x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9]

# Manual calculation of covariance
def manual_covariance(x, y):
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n
    cov = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n - 1)
    return cov

# Using NumPy for covariance
cov_matrix = np.cov(x, y, bias=False)  # bias=False -> sample covariance

# Display results
print("Manual Covariance:", manual_covariance(x, y))
print("NumPy Covariance Matrix:\n", cov_matrix)
print("Covariance from NumPy:", cov_matrix[0, 1])
```

```python
#10 Write a Python script to calculate the correlation coefficient between two datasets
import numpy as np

# Sample datasets
x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9]

# Manual calculation of Pearson correlation coefficient
def manual_correlation(x, y):
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n

    numerator = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n))
    denominator = (
        sum((x[i] - mean_x)**2 for i in range(n)) *
        sum((y[i] - mean_y)**2 for i in range(n))
    ) ** 0.5

    return numerator / denominator

# Using NumPy for correlation
corr_matrix = np.corrcoef(x, y)

# Display results
print("Manual Pearson Correlation Coefficient:", manual_correlation(x, y))
print("NumPy Correlation Matrix:\n", corr_matrix)
print("Correlation from NumPy:", corr_matrix[0, 1])
```

```python
#12 Implement and compare simple random sampling and systematic sampling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a synthetic population
population = pd.DataFrame({'ID': np.arange(1, 101), 'Value': np.random.normal(loc=50, scale=10, size=100)})

# Sample size
n = 10

# 1. Simple Random Sampling
srs_sample = population.sample(n=n, random_state=42)

# 2. Systematic Sampling
k = len(population) // n  # Sampling interval
start = np.random.randint(0, k)  # Random start between 0 and k-1
sys_indices = np.arange(start, len(population), k)
sys_sample = population.iloc[sys_indices[:n]]  # Ensure sample size = n

# Display samples
print("Simple Random Sample:\n", srs_sample)
print("\nSystematic Sample:\n", sys_sample)

# Compare samples visually
plt.figure(figsize=(10, 5))
sns.histplot(population['Value'], bins=15, color='gray', label='Population', kde=True)

sns.histplot(srs_sample['Value'], bins=10, color='blue', label='SRS', kde=True)
sns.histplot(sys_sample['Value'], bins=10, color='green', label='Systematic', kde=True)

plt.title("Sampling Methods Comparison")
plt.xlabel("Value")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#13 Calculate the mean, median, and mode of grouped data
import pandas as pd

# Define grouped data as a frequency distribution table
# Each row represents a class interval and its frequency
data = pd.DataFrame({
    'Lower': [0, 10, 20, 30, 40],
    'Upper': [10, 20, 30, 40, 50],
    'Frequency': [5, 8, 15, 16, 6]
})

# Calculate class midpoints
data['Midpoint'] = (data['Lower'] + data['Upper']) / 2

# Mean of grouped data
total_freq = data['Frequency'].sum()
mean = (data['Midpoint'] * data['Frequency']).sum() / total_freq

# Median of grouped data
cumulative_freq = data['Frequency'].cumsum()
median_class_index = cumulative_freq.searchsorted(total_freq / 2)
L = data.loc[median_class_index, 'Lower']
F = cumulative_freq.iloc[median_class_index - 1] if median_class_index > 0 else 0
f = data.loc[median_class_index, 'Frequency']
h = data.loc[median_class_index, 'Upper'] - data.loc[median_class_index, 'Lower']
median = L + ((total_freq / 2 - F) / f) * h

# Mode of grouped data
modal_class_index = data['Frequency'].idxmax()
L_modal = data.loc[modal_class_index, 'Lower']
f1 = data.loc[modal_class_index, 'Frequency']
f0 = data.loc[modal_class_index - 1, 'Frequency'] if modal_class_index > 0 else 0
f2 = data.loc[modal_class_index + 1, 'Frequency'] if modal_class_index < len(data) - 1 else 0
h_modal = data.loc[modal_class_index, 'Upper'] - data.loc[modal_class_index, 'Lower']
mode = L_modal + ((f1 - f0) / ((f1 - f0) + (f1 - f2))) * h_modal

# Output the results
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
```

```python
#14 Simulate data using Python and calculate its central tendency and dispersion
import numpy as np
import pandas as pd
from scipy import stats

# Simulate data (e.g., test scores from normal distribution)
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=100)  # mean=70, std=10, n=100

# Convert to pandas Series for easier handling
data_series = pd.Series(data)

# Central Tendency
mean = data_series.mean()
median = data_series.median()
mode = data_series.mode().iloc[0]  # use .iloc[0] to get the first mode if multiple

# Dispersion
data_range = data_series.max() - data_series.min()
variance = data_series.var()
std_dev = data_series.std()
iqr = stats.iqr(data_series)

# Display Results
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
print(f"Range: {data_range:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#24 Implement the Pearson and Spearman correlation coefficients for a dataset.
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Simulate a dataset with two variables (e.g., test scores)
np.random.seed(42)
x = np.random.normal(50, 10, 100)  # Random data for x
y = 0.5 * x + np.random.normal(0, 5, 100)  # y is linearly related to x, with some noise

# Create a DataFrame for easier handling
data = pd.DataFrame({'X': x, 'Y': y})

# Pearson Correlation
pearson_corr, _ = pearsonr(data['X'], data['Y'])

# Spearman Correlation
spearman_corr, _ = spearmanr(data['X'], data['Y'])

# Output the correlation coefficients
print(f"Pearson Correlation Coefficient: {pearson_corr:.2f}")
print(f"Spearman Correlation Coefficient: {spearman_corr:.2f}")
```
